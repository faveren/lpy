The costs of crude oil 1, 2, and 3 per barrel are \$2, \$3, and \$3, respectively. Formulate and solve the linear programming problem to maximize daily profit.

## Model Implementation

````python
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.R1 = Var(domain=NonNegativeReals)
model.R2 = Var(domain=NonNegativeReals)
model.R3 = Var(domain=NonNegativeReals)

model.S1 = Var(domain=NonNegativeReals)
model.S2 = Var(domain=NonNegativeReals)
model.S3 = Var(domain=NonNegativeReals)

model.E1 = Var(domain=NonNegativeReals)
model.E2 = Var(domain=NonNegativeReals)
model.E3 = Var(domain=NonNegativeReals)

# declare objective
model.lucro = Objective(
    expr =
        5 * (model.R1 + model.R2 + model.R3)
      + 7 * (model.S1 + model.S2 + model.S3)
      + 8 * (model.E1 + model.E2 + model.E3)
      - 2 * (model.R1 + model.S1 + model.E1)
      - 3 * (model.R2 + model.S2 + model.E2)
      - 3 * (model.R3 + model.S3 + model.E3),
    sense = maximize
)

# declare constraints
model.regular_oil1 = Constraint(
    expr = model.R1 <= 0.7 * (model.R1 + model.R2 + model.R3)
)

model.super_oil1 = Constraint(
    expr = model.S1 <= 0.5 * (model.S1 + model.S2 + model.S3)
)

model.super_oil2 = Constraint(
    expr = model.S2 >= 0.1 * (model.S1 + model.S2 + model.S3)
)

model.extra_oil2 = Constraint(
    expr = model.E2 <= 0.5 * (model.E1 + model.E2 + model.E3)
)

model.extra_oil3 = Constraint(
    expr = model.E3 >= 0.4 * (model.E1 + model.E2 + model.E3)
)

model.demandaRegular = Constraint(
    expr = model.R1 + model.R2 + model.R3 >= 5000
)

model.demandaSuper = Constraint(
    expr = model.S1 + model.S2 + model.S3 >= 3000
)

model.demandaExtra = Constraint(
    expr = model.E1 + model.E2 + model.E3 >= 3000
)

model.oil1 = Constraint(
    expr = model.R1 + model.S1 + model.E1 <= 10000
)

model.oil2 = Constraint(
    expr = model.R2 + model.S2 + model.E2 <= 8000
)

model.oil3 = Constraint(
    expr = model.R3 + model.S3 + model.E3 <= 7000
)

model.capacidadeTotal = Constraint(
    expr = model.R1 + model.R2 + model.R3
         + model.S1 + model.S2 + model.S3
         + model.E1 + model.E2 + model.E3 <= 20000
)

model.pprint()
````

# Problem 2

CPS Investment Services must develop an investment portfolio for a new customer. Initially, the new customer would like to restrict portfolio to three stocks as shown below:

|Stock|Price per Share ($)|Estimated Annual Return (%)|
|--|--|--|
|ACT cable|50|6|
|CTR retail|35|10|
|PCL petroleum|120|8|

The customer wants to invest $80,000 and established two investment
goals.
Goal 1: Obtain an annual return of at least 9%.
Goal 2: Limit the investment in CTR retail, the riskier investment, to no more than 50% of the total investment.

Formulate a mathematical programming model for the CPS investment problem and solve it.

# Solution 2

# Model

X<sub>1</sub> is the amount invested in ACT cable  

X<sub>2</sub> is the amount invested in CTR retail  

X<sub>3</sub> is the amount invested in PCL petroleum  

max 0.06X<sub>1</sub> + 0.10X<sub>2</sub> + 0.08X<sub>3</sub>  

s.t.  

X<sub>1</sub> + X<sub>2</sub> + X<sub>3</sub> = 80000  

0.06X<sub>1</sub> + 0.10X<sub>2</sub> + 0.08X<sub>3</sub> ≥ 7200  

X<sub>2</sub> ≤ 40000  

X<sub>1</sub>, X<sub>2</sub>, X<sub>3</sub> ≥ 0

# Model Implementation

````python
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.p1 = Var(domain=NonNegativeReals)
model.p2 = Var(domain=NonNegativeReals)
model.p3 = Var(domain=NonNegativeReals)

# declare objective
model.profit = Objective(
    expr = 0.06*model.p1 + 0.10*model.p2 + 0.08*model.p3,
    sense = maximize
)

# declare constraints
model.totalInvestment = Constraint(
    expr = model.p1 + model.p2 + model.p3 == 80000
)

model.minimumReturn = Constraint(
    expr = 0.06*model.p1 + 0.10*model.p2 + 0.08*model.p3 >= 7200
)

model.ctrLimit = Constraint(
    expr = model.p2 <= 40000
)

model.pprint()
````

# Solving

````python
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()

# display solution
print('\nProfit = ', model.profit())

print('\nDecision Variables')
print('P1 = ', model.p1())
print('P2 = ', model.p2())
print('P3 = ', model.p3())

print('\nConstraints')
print('Total Investment = ', model.totalInvestment())
print('Minimum Return = ', model.minimumReturn())
print('CTR Limit = ', model.ctrLimit())
````

# Problem 3

A popular product is produced at three plants (PL1, PL2, PL3) and is required to ship to three warehouses: WH1, WH2, and WH3. The supply from each plant, the delivery required in each warehouse, and the unit transportation cost from each plant to each warehouse is shown below:

|Plant|WH1|WH2|WH3|Plant Capacity|
|--|--|--|--|--|
|PL1|20|15|25|400|
|PL2|10|12|8|500|
|PL3|14|20|12|200|
|Warehouse demand|300|200|600||

Develop an LP model for minimizing the transportation cost while not violating the supply and demand constraints.

# Solution 3

# Model

X<sub>11</sub> is the number of units shipped from PL1 to WH1  

X<sub>12</sub> is the number of units shipped from PL1 to WH2  

X<sub>13</sub> is the number of units shipped from PL1 to WH3  

X<sub>21</sub> is the number of units shipped from PL2 to WH1  

X<sub>22</sub> is the number of units shipped from PL2 to WH2  

X<sub>23</sub> is the number of units shipped from PL2 to WH3  

X<sub>31</sub> is the number of units shipped from PL3 to WH1  

X<sub>32</sub> is the number of units shipped from PL3 to WH2  

X<sub>33</sub> is the number of units shipped from PL3 to WH3  

min 20X<sub>11</sub> + 15X<sub>12</sub> + 25X<sub>13</sub> + 10X<sub>21</sub> + 12X<sub>22</sub> + 8X<sub>23</sub> + 14X<sub>31</sub> + 20X<sub>32</sub> + 12X<sub>33</sub>  

s.t.  

X<sub>11</sub> + X<sub>12</sub> + X<sub>13</sub> ≤ 400  

X<sub>21</sub> + X<sub>22</sub> + X<sub>23</sub> ≤ 500  

X<sub>31</sub> + X<sub>32</sub> + X<sub>33</sub> ≤ 200

X<sub>13</sub> + X<sub>23</sub> + X<sub>33</sub> = 600  

X<sub>11</sub>, X<sub>12</sub>, X<sub>13</sub>, X<sub>21</sub>, X<sub>22</sub>, X<sub>23</sub>, X<sub>31</sub>, X<sub>32</sub>, X<sub>33</sub> ≥ 0

# Model Implementation

````python
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.p11 = Var(domain=NonNegativeReals)
model.p12 = Var(domain=NonNegativeReals)
model.p13 = Var(domain=NonNegativeReals)

model.p21 = Var(domain=NonNegativeReals)
model.p22 = Var(domain=NonNegativeReals)
model.p23 = Var(domain=NonNegativeReals)

model.p31 = Var(domain=NonNegativeReals)
model.p32 = Var(domain=NonNegativeReals)
model.p33 = Var(domain=NonNegativeReals)

# declare objective
model.cost = Objective(
    expr =
        20*model.p11 + 15*model.p12 + 25*model.p13
      + 10*model.p21 + 12*model.p22 + 8*model.p23
      + 14*model.p31 + 20*model.p32 + 12*model.p33,
    sense=minimize
)

# declare constraints
model.plant1 = Constraint(
    expr = model.p11 + model.p12 + model.p13 <= 400
)

model.plant2 = Constraint(
    expr = model.p21 + model.p22 + model.p23 <= 500
)

model.plant3 = Constraint(
    expr = model.p31 + model.p32 + model.p33 <= 200
)

model.warehouse1 = Constraint(
    expr = model.p11 + model.p21 + model.p31 == 300
)

model.warehouse2 = Constraint(
    expr = model.p12 + model.p22 + model.p32 == 200
)

model.warehouse3 = Constraint(
    expr = model.p13 + model.p23 + model.p33 == 600
)

model.pprint()
````

# Solving

````python
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()

# display solution
print('\nCost = ', model.cost())

print('\nDecision Variables')
print('P11 = ', model.p11())
print('P12 = ', model.p12())
print('P13 = ', model.p13())
print('P21 = ', model.p21())
print('P22 = ', model.p22())
print('P23 = ', model.p23())
print('P31 = ', model.p31())
print('P32 = ', model.p32())
print('P33 = ', model.p33())

print('\nConstraints')
print('Plant 1 = ', model.plant1())
print('Plant 2 = ', model.plant2())
print('Plant 3 = ', model.plant3())
print('Warehouse 1 = ', model.warehouse1())
print('Warehouse 2 = ', model.warehouse2())
print('Warehouse 3 = ', model.warehouse3())
````

# Problem 4

A certain paint can be produced using four different production processes (P1, P2, P3, and P4). The processing cost of each liter in any of the four available processes, the maximum capacity of each process, and the setup costs are given below:

|Process|Setup Cost (\$)|Processing Cost (\$/L)|Capacity (L)|
|--|--|--|--|
|P1|5,000|0.60|20,000|
|P2|6,000|0.50|15,000|
|P3|10,000|0.40|40,000|
|P4|6,000|0.40|25,000|

Assume that a daily demand of 45,000 L must be fulfilled. Formulate the problem as a programming model to determine the daily production schedule in order to minimize total costs.

# Solution 4

# Model

X<sub>1</sub> is the number of liters produced using process P1  

X<sub>2</sub> is the number of liters produced using process P2  

X<sub>3</sub> is the number of liters produced using process P3  

X<sub>4</sub> is the number of liters produced using process P4  

Y<sub>1</sub> is 1 if process P1 is used, and 0 otherwise  

Y<sub>2</sub> is 1 if process P2 is used, and 0 otherwise  

Y<sub>3</sub> is 1 if process P3 is used, and 0 otherwise  

Y<sub>4</sub> is 1 if process P4 is used, and 0 otherwise  

min 5000Y<sub>1</sub> + 6000Y<sub>2</sub> + 10000Y<sub>3</sub> + 6000Y<sub>4</sub> + 0.60X<sub>1</sub> + 0.50X<sub>2</sub> + 0.40X<sub>3</sub> + 0.40X<sub>4</sub>  

s.t.  

X<sub>1</sub> + X<sub>2</sub> + X<sub>3</sub> + X<sub>4</sub> = 45000  

X<sub>1</sub> ≤ 20000Y<sub>1</sub>  

X<sub>2</sub> ≤ 15000Y<sub>2</sub>  

X<sub>3</sub> ≤ 40000Y<sub>3</sub>  

X<sub>4</sub> ≤ 25000Y<sub>4</sub>  

X<sub>1</sub>, X<sub>2</sub>, X<sub>3</sub>, X<sub>4</sub> ≥ 0  

Y<sub>1</sub>, Y<sub>2</sub>, Y<sub>3</sub>, Y<sub>4</sub> ∈ {0,1}

# Model Implementation

````python
from pyomo.environ import *

# create a model
model = ConcreteModel()

# declare decision variables
model.x1 = Var(domain=NonNegativeReals)
model.x2 = Var(domain=NonNegativeReals)
model.x3 = Var(domain=NonNegativeReals)
model.x4 = Var(domain=NonNegativeReals)

model.y1 = Var(domain=Binary)
model.y2 = Var(domain=Binary)
model.y3 = Var(domain=Binary)
model.y4 = Var(domain=Binary)

# declare objective
model.cost = Objective(
    expr = 5000*model.y1 + 6000*model.y2 + 10000*model.y3 + 6000*model.y4
         + 0.60*model.x1 + 0.50*model.x2 + 0.40*model.x3 + 0.40*model.x4,
    sense = minimize
)

# declare constraints
model.demand = Constraint(
    expr = model.x1 + model.x2 + model.x3 + model.x4 == 45000
)

model.capacityP1 = Constraint(
    expr = model.x1 <= 20000*model.y1
)

model.capacityP2 = Constraint(
    expr = model.x2 <= 15000*model.y2
)

model.capacityP3 = Constraint(
    expr = model.x3 <= 40000*model.y3
)

model.capacityP4 = Constraint(
    expr = model.x4 <= 25000*model.y4
)

model.pprint()
````

# Solving

````python
SolverFactory('glpk', executable='/usr/bin/glpsol').solve(model).write()

# display solution
print('\nCost = ', model.cost())

print('\nDecision Variables')
print('X1 = ', model.x1())
print('X2 = ', model.x2())
print('X3 = ', model.x3())
print('X4 = ', model.x4())
print('Y1 = ', model.y1())
print('Y2 = ', model.y2())
print('Y3 = ', model.y3())
print('Y4 = ', model.y4())

print('\nConstraints')
print('Demand = ', model.demand())
print('Capacity P1 = ', model.capacityP1())
print('Capacity P2 = ', model.capacityP2())
print('Capacity P3 = ', model.capacityP3())
print('Capacity P4 = ', model.capacityP4())
````

Linear Programming (LP) problems can be classified according to whether feasible solutions exist and how many optimal solutions they have. In general, there are four main classes: **unique optimal solution, multiple optimal solutions, infeasible solution, and unbounded solution**.

A **unique optimal solution** occurs when the feasible region exists and the objective function reaches its best value at only one point. For example, a company may find that only one product mix gives the maximum profit.

A **multiple optimal solution** occurs when two or more feasible solutions give the same optimal value. This usually happens when the objective function is parallel to a boundary of the feasible region. For example, different production combinations may yield the same total profit.

An **infeasible solution** means there is no feasible region because the constraints conflict with one another. For example, a model may require production to be at least 100 units while available resources allow at most 50 units.

An **unbounded solution** occurs when the feasible region exists, but the objective function can increase forever in a maximization problem or decrease forever in a minimization problem. This often means an important constraint is missing.

These four classes help us understand whether an LP problem has one best answer, several best answers, no feasible answer, or no finite optimal answer.
